```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

# Set the mobile number
mobile_no = "+918870073991"

# Set the frequency of the operation
freq = 5  # Change this to the number of times you want to execute

# Set up the WebDriver (Chrome in this case)
# Make sure to replace 'path/to/chromedriver' with the actual path to your ChromeDriver
driver_path = 'C:/Program Files/Google/Chrome/Application/chrome.exe'
browser = webdriver.Chrome(driver_path)

try:
    for i in range(freq):
        # Open Flipkart login page
        browser.get('https://www.flipkart.com/account/login?ret=/')

        # Find the mobile number input field using its XPath and enter the mobile number
        number = browser.find_element(By.XPATH, '//*[@id="container"]/div/div[3]/div/div[2]/div/form/div[1]/input')
        number.send_keys(mobile_no)

        # Find the "Forgot?" link and click on it
        forgot = browser.find_element(By.LINK_TEXT, 'Forgot?')
        forgot.click()

        # Wait for 2 seconds before sending the next request
        time.sleep(2)

finally:
    # Close the browser
    browser.quit()
```
