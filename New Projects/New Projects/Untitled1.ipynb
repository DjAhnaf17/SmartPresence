```python
import os
from datetime import datetime
from twilio.rest import Client

# Twilio credentials
account_sid = 'ACcee8f326e33c1bd9f1b7af26cfc46081'
auth_token = 'ad8edfd846752e8d81b6c68f16a0321e'
client = Client(account_sid, auth_token)

# Contacts to send the message to
contacts = ["+919500629805", "+918870073991","+917695970785"]

# Get current date and time
current_date_time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

# Message content
for i in contacts:
    for j in range(1,10):
        message_body = (
            f"⚠️ Warning from Admin ⚠️\n\n"
            f"This is to notify you that your attendance is below 90%.\n"
            f"Date and Time of Notification: {current_date_time}\n"
            f"Please take immediate action to ensure compliance with attendance policies.\n"
            f"Contact the admin office if you have any questions."
        )
        
        # Send message using Twilio
        message = client.messages.create(
            body=message_body,
            from_="+12298140498",  # Your Twilio phone number
            to=i,
        )
    
        print(f"Message sent to {i}: {message.body}")
```
